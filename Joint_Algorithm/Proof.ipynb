### 01 problem

```python
import numpy as np

import sympy

from sympy import symbols

from sympy import nonlinsolve

import matplotlib.pyplot as plt

from matplotlib.ticker import MultipleLocator, FormatStrFormatter
```

\begin{equation}
H^{\textrm{Global}} = H^{\textrm{Dispatch}} \Longleftrightarrow (\frac{nk+1-k}{nk+1})^{n}=\frac{1}{k+1}\Longleftrightarrow(k+1)(\frac{nk+1-k}{nk+1})^{n}=1
\end{equation}

### 02 Solver

```python
# solver

def prob(n):
    
    k = symbols('k',real=True)
    
    solution=list()
    
    for s in list(nonlinsolve([(k+1)*((n*k+1-k)/(n*k+1))**n-1],[k])):
        
        if s[0].is_real:
            
            if s[0]>0:
            
                solution.append(s[0])
    

    return  solution

# solution

k_opt={}

for n in range(2,25,1):

    solution=prob(n)
    
    k_opt[n]=solution[0]
    
    print('n is ',n)
    
    print('current solution is ',float(solution[0]))
    
    print('\n'*2)
    
    print('*'*100)
```

```python
# figures

figsize = 10,5

font_title = {'family' : 'Arial','weight' : 'normal','size': 30}

font_label = {'family' : 'Arial','weight' : 'normal','size': 20}

font_legend = {'family' : 'Arial','weight' : 'normal','size': 20}


figure, axs = plt.subplots(figsize=figsize)



labels = axs.get_xticklabels() + axs.get_yticklabels()

[label.set_fontsize(15) for label in labels]

[label.set_fontname('Arial') for label in labels]

axs.set_ylabel('k*',font_label)

axs.set_xlabel('n',font_label)

axs.grid()
    

axs.plot(list(k_opt.keys()),list(k_opt.values()),'b--',linewidth=2,marker='x',markersize=10,markeredgecolor='r')

xmajorLocator   = MultipleLocator(1)

axs.xaxis.set_major_locator(xmajorLocator)

current_fig = plt.gcf() # 'get current figure'

current_fig.savefig('K_Equality.eps', format='eps', dpi=1000, bbox_inches = 'tight')
        
plt.tight_layout()

plt.show()
```
